##### slice 透過 [ : ] 進行分割
`[n:m]` 走訪 n 到 m - 1 的元素

```python
# 初始化一個字串
data = "Hello,World!"
print(data)
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
data = "Hello,World!"
print(data)
print(data[0:5])
```

```python
# 從頭取到 5（不包含第 5 個元素，實際為 0 ~ 4）
data = "Hello,World!"
print(data)
print(data[:5])
```

```python
# 從 6 取到尾
data = "Hello,World!"
print(data)
print(data[6:])
```

```python
# 從 7 取到 9（不包含第 9 個元素，實際為 7 ~ 8）
data = "Hello,World!"
print(data[7:9])
```

##### 使用負號

```python
# 從倒數第 10 取到倒數第 7（不包含倒數第 7 的元素）
data = "Hello,World!"
print(data[-10:-7])
```

```python
# 從倒數第 6 取到尾
data = "Hello,World!"
print(data[-6:])
```

```python
# 從頭取到倒數第 7（不包含倒數第 7 的元素）
data = "Hello,World!"
print(data[:-7])
```

```python
# 也可以使用在 list
nums = [i for i in range(1, 16)]
print(nums)
print(nums[-5:])
print(nums[:10]) # 看前 10 筆的資料
```

```python
# 實做練習 將每個單字改成大寫開頭
texts = "hello world! hello john"
# print(text.title())

words = texts.split(' ')
print(words)
result = ""
for word in words:
    result += word[0].upper() + word[1:] + " "

result = result.strip() # 去除空格
print(result)
```

```python
# ChatGPT
words = texts.split()
print(words)

result = []

for word in words:
    new_word = word[0].upper() + word[1:]
    result.append(new_word)

print(" ".join(result))
```
