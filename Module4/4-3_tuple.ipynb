```python
# tuple 初始化: 第一種
data = ("Alex", "Allen")
print(data)
print(type(data))
```

```python
# tuple 初始化: 第二種
data = "Alex", "Allen", "Alice"
print(data)
print(type(data))
```

```python
# 透過指定索引輸出值
data = ("Alex", "Allen")
print(data[0])
```

```python
# Tuple 「解包」的功能用法
name, age = ("Alex", 30)
print(name)
print(age)

name, age = "Alex", 30
print(name)
print(age)
```

```python
# tuple 不可以修改指定索引的值
data = (1, 2, 3)
data[0] = 100
```

```python
# 用 for 迴圈逐一輸出資料
data = (1, 2, 3)
for i in data:
    print(i)
```

```python
# 用 len 計算 tuple 資料個數
data = (1, 2, 3)
print(len(data))
```
