|語法|說明|
|:---:|:---:|
| `%s` |插入 `字串` 的值到字串中|
| `%d` |插入 `整數` 的值到字串中|
| `` |插入  的值到字串中|
| |輸出  (類似跳脫字元)|

```python
# 多組文字

word1 = "Hello"
word2 = "World"

msg = "[msg], %s, %s!" % (word1, word2)
print(msg)
```

```python
# 整數
price = 50

msg = "蘋果 %d 元" % (price)
print(msg)
print(type(msg))
```

```python
# 文字與整數
name = "Ollie"
age = 18

result = "%s is %d years old." % (name, age)
print(result)
print(type(result))
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % ("Hello")
print(msg)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = "[%-10s]" % ("Hello")
print("="*13)
print(msg)
print("="*13)
```

```python
# 指定浮點數位數
# 會四捨五入
result = "%.3f" % (3.1415926)
print(result)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
result = "%.3s" % ("Hello")
print(result)
```

```python
# 空白補 0
# 不會四捨五入
result = "[%06.2f]" % (3.1415926)
print(result)
```
