```python
# 字串變數初始化
data = "Hello World!"
print(data)
print(data[0]) # Python 由 0 起算
```

```python
# 字串是可以相加相乘的
data = "Hell" + "o"*5 + "," + "World!"
print("=" * 20)
print(data)
print("=" * 20)
print(len(data)) # len()
```

### Method

```python
# 重新賦值，字串仍是不可變的
name = "Alex"
name = "Allen"
print(name)
```

```python
# 替換字串
'''
str.replace(<old>, <new>)
將 string 中的 str1 替換成 str2
'''
name =  "Alex"

# 會Return一個新的變數，而不是改變原本的變數，所以要用一個變數去接回傳的值。
new_name = name.replace("ex",  "len")
print(name)
print(new_name)
```

```python
# 去除兩側空格 .strip()

'''

去除字串 string 左、右兩邊的空格
'''
msg = " Hello,Wolrd "

result = msg.strip()
print(result)
```

```python
# 去除指定字串 .strip("指定字串")

msg = "_Hello,Wolrd_"

result = msg.strip("_")
print(result)
```

```python
# 字串小寫變大寫
'''

將字串 string 裡的字母全部改成小寫
'''
msg = "HELLO!"
result = msg.lower()
print(result)


'''

將字串 string 裡的字母全部改成大寫
'''
name = "ollie"
result = name.upper()
print(result)
```

```python
# 字串分割成 list -> str.split()
greeting = "hello world! hello John"

word_list = greeting.split(" ")
print(word_list)

# 也可以指定字串做分割
greeting = "hello world! hello John"

word_list = greeting.split("!")
print(word_list)
```
