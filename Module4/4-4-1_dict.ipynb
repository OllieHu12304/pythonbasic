```python
# 資料較多時，建議的寫法
fruit_price = {
    "apple": 40,
    "orange": 50,
    "tomato": 100
}

print(fruit_price)
```

```python
# 印出 蘋果的價格
print(fruit_price["apple"])
```

```python
# 修改 橘子的價格 (使用中括號[])
fruit_price["orange"] = 150
print(fruit_price)
```

```python
# 新增 kiwi
fruit_price['kiwi'] = 30 #原本沒有該key值，就新增
print(fruit_price)
```

```python
# 刪除 番茄 dic.pop(<key>)
fruit_price.pop("tomato")
print(fruit_price)
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name": "Ollie",
    "age":30,
    "info": {
        "nickname": "Oli",
        "role": "Programmer",
        "phones": ["0953330624", "07-3334470"]
    }
}
print(user_info)
```

```python
# 取得所有 phone_number
print(user_info["info"]["phones"][0])
print(user_info["info"]["phones"][1])
user_phone1 = user_info["info"]["phones"][0]
print(user_phone1)
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    'apple':{
        "price": 50,
        "quantity": 100,
        "color": ["green", "red"]
        },
        "orange": {
            "price": 100,
            "quantity": 50,
            "color": ["orange"]
        },
        "tomato": {
            "price": 100,
            "quantity": 50,
            "color": ["red"]
        }
}
# 記得只要是「值」要加上中括號，否則其資料型別會是str，在索引時會被逐一'o', 'r', 'a', 'n', 'g', 'e'
print(fruit_dict)
```

##### dict 的合併

```python
# dict 的合併 dict.update()
data = {
    "banana": {
        "price": 20,
        "quantity": 50,
        "color": ["yellow"]
    }
}

fruit_dict.update(data)
print(fruit_dict)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict)  # 注意順序會被升冪排列
print("=" * 80)
pprint(fruit_dict, sort_dicts = False) #若不想被自動排序
```

```python
# 也可以寫成
import pprint
pprint.pprint(fruit_dict)
# 第一個 pprint 是模組名稱，第二個 pprint 是模組裡的函式
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''
fruit_dict = {
    'apple':{
        "price": 50,
        "quantity": 100,
        "color": ["green", "red"]
        },
        "orange": {
            "price": 100,
            "quantity": 50,
            "color": ["orange"]
        },
        "tomato": {
            "price": 100,
            "quantity": 50,
            "color": ["red"]
        }
}

print(fruit_dict['apple']['color']) # 確定資料型態是 list
fruit_dict['apple']['color'].append("yellow")
fruit_dict['tomato']['color'].append("green")

pprint(fruit_dict, sort_dicts=False)
```
