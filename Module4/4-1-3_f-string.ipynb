```python
# f-string (又稱作 formatted string literals, version >= 3.6)
msg = "Hello World"

print(f"[msg] {msg}!!")
```

```python
name = "Ollie"
age = 18

print(f"{name} is {age} years old.")
```

```python
# 靠左對齊、靠右對齊
# 預設為不足補空白
data = "Hello"

print(f"[{data:<10}]")
print(f"[{data:>10}]")
```

```python
# 靠左對齊、靠右對齊，字數不足，補「_」
# (沒寫補什麼，預設空格)
data = "Hello"

print(f"[{data:_<10}]")
print(f"[{data:X<10}]")
```

```python
# 置中對齊
data = "Hello"

print(f"[{data:^10}]")
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
data = "Hello"

print(f"[{data:.2}]")
```

```python
num = 3.14256
print(f"{num:.2f}")
```

```python
# 金額
total = 1506583
print(f"{total:,}")
```

```python
# 題目:使用f String，想要將身分證字號回傳"A123******"

user_id = "A123456789"
print(f"{user_id:.4}" + "******")

print(f"{user_id[0]}{user_id[1]}{user_id[2]}{user_id[3]}******")

print(f"{user_id:*<10.4}")
```
