```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['greem', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1 -> 只回傳 key 
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2 -> dict.values() 只回傳 value
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料3 -> dict.items() key 與 value 都回傳
for value in fruit_dict.items():
    print(value)
```

```python
for key, value in fruit_dict.items():
    for color in value['color']:
        print(f"{key} 有 {color}")
```

`.join()` #注意，只能連接字串(str)<br>
工作只有一件事：把多個「字串」串接成一個字串。

```python
# apple 有 green, red
for key, value in fruit_dict.items():
    print(f"{key} 有 {', '.join(value['color'])}")
```

```python
# 做法 2
for key, value in fruit_dict.items():
    result = ', '.join(value['color'])
    print(f"{key} 有 {result}")
```

`.get()`<br>
安全地取得某個 key 對應的值，若不存在回報 `None` 而不會報錯。

```python
# 取得 dict 的 value 另一種方式
print(fruit_dict)

# print(fruit_dict['kiwi']) # 沒有該 key 會報錯

print(fruit_dict.get('kiwi'))

if fruit_dict.get('kiwi') is None:
    print(f'kiwi 不在 fruit_dict 裡')

# .get()後也可以接著輸入「當索引不到對應key值」時回傳的訊息
print(fruit_dict.get('kiwi', '沒有該值'))
```

```python
# 資料處理
# 看到 list [] ， 第一反應就是跑 for 迴圈
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
# 難！
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}

# 此處 fruit 是 key ， color 是 value


# 初始化一個空字典，用來存放最終分類與整合後的資料
result = {}

# 第一層迴圈：遍歷列表，依序取出裡面的每一個字典物件
for fruit_color in fruit_color_list:
    
    # 第二層迴圈：解構目前取出的字典，將其鍵（Key）賦值給 fruit，值（Value）賦值給 color
    for fruit, color in fruit_color.items():
        
        # 判斷式：檢查 result 字典中是否「還沒有」存在 fruit 這個鍵
        if result.get(fruit) == None: 
            
            # 若不存在：在 result 中建立該 fruit 鍵，並將值初始化為一個「包含當前顏色的列表」
            result[fruit] = [color]
            
        else:
            # 若已存在：代表 result[fruit] 目前已經是一個列表，直接使用 .append() 將新顏色加入該列表末尾
            result[fruit].append(color)

# 輸出最終整合完成的字典結果
print(result)
```

##### `defaultdict(<type>)`

```python
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
from collections import defaultdict

# 以int型態的預設值 0 來初始化
result = defaultdict(list)

print(result['apple'])
print(result)
```

```python
from collections import defaultdict

fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}
]

result = defaultdict(list)

for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        result[fruit].append(color)

result = dict(result) #強制轉型成一般的 dict
print(result)
```
