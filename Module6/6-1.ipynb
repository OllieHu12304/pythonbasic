```python
# 類別名稱
class Human:
    # 類別屬性 (Class Attribute)
    hands = 2
    legs = 2

# 取得類別屬性的值
print(Human.hands) 
print(Human.legs)
```

```python
# 類別名稱
class Car:
    # 類別屬性 (Class Attribute)
    wheels = 4
    def __init__(self): # 建立物件預設第一時間要執行的函式
        print("已建立一個物件")
# 方法 (Method)

print(Car.wheels)

car1 = Car() # 建立物件
```
