##### try...except...

```python
print(y)
```

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try:
    x = 10
    print(y)
    print(x) # 上面有抱錯，下面都不執行
except: # 捕捉所有的錯誤
    print("發生錯誤")
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(x) 
# except Exception as e: # 錯誤用法
#     print(f'發生錯誤: {e}')
except NameError: # 捕捉特定錯誤類別
    print("發生 NameError")
except TypeError: 
    print("發生 TypeError")
except Exception as e: # 一律放在最下面
    print(f'發生錯誤: {e}')
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(x) 
    # as -> 把 Exception 產生的資料賦值給 e
except Exception as e: # 捕捉大部分錯誤類別
    print(f"發生錯誤: {e}")
```

```python
# 多層例外處理:
try:
    x = 10
    try:
        print(y)
    except NameError as e:
        print(f'發生錯誤: {e}')
    print(x)
except Exception as e:
    print(e)
```
