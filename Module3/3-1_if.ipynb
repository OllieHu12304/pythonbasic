條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
age = 20

if age == 20:
    print("條件成立")
```

##### if...else...

```python
score = 50

if score >= 60:
    print("及格")
else: # if 敘述為 False 就執行 else
    print("不及格")
```

```python
age = 25

if age == 20:
    print("條件成立")
else:
    print("條件不成立")
```

##### if...elif... (else)

```python
score = 95
if score == 100:
    print("算你狠==!")
elif score >= 90:
    print("就還行吧")
elif score >= 60:
    print("加油好嗎")
else:
    print("嫩")
```
