| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
# range(m-1)
# 記得沒給起點是由0起算
for i in range(5):
    print(i, end=", ") 
print()
print("Hello\n\nWorld") 

# 「 end = ", " 」 可以避免換行
# end 表示 「這次 print 結束後，要接什麼字元」。
# print()、「 end = "\n" 」 可以強置換行
```

```python
# 當變數不會在後面被使用，將其設為"_"是一種默契
for _ in range(3):
    print("Hello World")
```

```python
# range(n, m-1)
# 1 ~ 10 的數字
for n in range(1, 11):
    print(n, end=", ")
```

```python
# range(n, m-1, step)
for i in range(1, 16, 2):
    print(i, end=", ")
```

```python
nums = [1, 2, 3, 10, 7]

for num in nums:
    print(num, end=", ")
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
a = 6
n = 7

for a in range(a, a + 2 * n, 2):
    if a % 2 == 0:
        print(a, end=", ")
    else:
        print(a+1, end=", ")
```

```python
a = 4
n = 5

for i in range(n):
    if a % 2 ==0:
        # 迴圈第一次執行 4 + 0*2
        # 迴圈第一次執行 4 + 1*2
        # 迴圈第一次執行 4 + 2*2
        # 迴圈第一次執行 4 + 3*2
        # 迴圈第一次執行 4 + 4*2
        print(a + i * 2,end=", ")
    else:
        print(a + i * 2 + 1,end=", ")
```

```python
a = 5
n = 3

if a % 2 != 0:
    a += 1

for i in range(n):
    print(a + i*2, end=" ")
```

```python
# 嘗試產出
"""
* * * * *
 * * * *
  * * *
   * *
    *
"""
for i in range(5):
    print(" " * i + "* " * (5 - i))

    

# 嘗試產出
"""
    *
   * *
  * * *
 * * * *
* * * * *
"""
for i in range(5):
    print(" " * (4-i) + "* " * (i+1))
```
