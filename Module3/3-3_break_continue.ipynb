##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
for char in "contents":
    if char == "t":
        print("遇到字母 t 結束整個迴圈執行")
        break
    print(char) # break 條件成立後迴圈便終止
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
for char in "contents":
    if char == "t":
        print("遇到字母 t 跳過 continue 以下程式碼 進行下一次迴圈")
        continue
    print(char)
```

### 小試身手
計算 1 到 100 之間的和，如果值超過500就停止執行

```python
total = 0
for i in range(1, 101):
    total += i
    # print(i, end=", ")
    if total >= 500:
        break

print(total)
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python
total = 0

for i in range(1, 101):
    if i % 2 == 0 or i % 5 == 0:
        continue
    total += i
print(total)
```
