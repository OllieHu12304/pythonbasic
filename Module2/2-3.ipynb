整數 (int)、浮點數 (float)

```python
num = 20
float_num = 12.5
print(num)
print(float_num)

# 複製該行程式碼 → Alt + Shift + 方向鍵
# 移動該行程式碼 → Alt + 方向鍵
# 打開/關閉終端機 → Ctrl + `
# 移動到終端機 → Ctrl + `
# 移動到某個 window → Ctrl + 1 / Ctrl + 2 / Ctrl + 3
# 關閉window → Ctrl + W
# 同時改多個變數 → Ctrl + D
```

布林值 (boolean)

```python
parse_ok = True
parse_not_ok = False

print(parse_ok)
print(parse_not_ok)
print(type(parse_ok))
```

字串 (string)

```python
name = "Ollie"
print(name)
```

```python
data = "True"
print(data)
# 檢查資料型態 type()
print(type(data))
```

列表 (list)

```python
num_list = [1, 2, 3, 4, 5]
print(num_list)
print(type(num_list))
```

元組 (tuple)

```python
num_tuple = (1, 2, 3, 4, 5)
print(num_tuple)
print(type(num_tuple))
```

集合 (set)

```python
num_set = {1, 2, 3, 3, 5}
# set不會有重複的值
print(num_set)
print(type(num_set))
```

字典 (dict)

```python
user_info = {"name": "Ollie", "age": 20}
print(user_info)
print(type(user_info))
```

```python
# 空的List
result = []
print(result)
print(type(result))
```

```python
# 創建一個空的 dict
result = {}
print(result)
print(type(result))
```

```python
# 創建一個空的 set
empy_set = set()
print(empy_set)
print(type(empy_set))
```
