算術運算子

```python
a = 10
b = 3 

print(a / b)

# a 的2次方
print(a ** 2)

# 相除取餘數
print(a % b)

# 相除取商
print(a // b)

# 判斷 a 是否為偶數
print(a % 2 )
```

比較運算子 (用 if 判斷)

```python
a = 5
b = 10

print(a == b) # 沒有相等 回傳 False
print(a != b) #不等於 回傳 True
print(a >= b) #沒有大於等於 回傳 False
```

```python
score = 50

# 如果分數小於 60
if score < 60:
    score = score + 10 # score += 10
    # Tab縮排
    # 反縮排 Shift + Tab
    print("加分")

print("最終分數")
print(score)
```

邏輯運算子 and / or / not

```python
role = "VIP"
age = 15 

# or 只要其中一個條件有符合，就執行區塊內的程式碼
if role =="VIP" or age > 18:
    print("Or: 符合資格")

if role == "VIP" and age > 18:
    print("And:符合資格") # and 兩個條件都要符合
```

```python
parse_not = False

# if not False
if not parse_not:
    print("Not: False 變成 True")
```

成員運算子 in / not in

```python
student_ids = [1, 2, 3, 4, 5]
if 3 in student_ids:
    print("輸出成績單")

if 7 not in student_ids:
    print("7 號不再名單內")
```

身分運算子 is / is not

```python
x = 5
y = x

# 不是拿來判斷 兩個變數 是否相等
if y is x:
    print("在同一個記憶體位置")
print(y)
```

```python
x = [1, 2, 3]
y = [1, 2, 3]

if x is y:
    print("兩值相等")

# 查看記憶體位置
print(id(x))
print(id(y))
```
