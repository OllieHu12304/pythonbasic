### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
| `json.dump(<list/dict>, file, indent=<num>)` |物件轉JSON並寫入文件|
| `json.dumps(<list/dict>, indent=<num>)` |物件轉JSON字串|
| `json.load(<file>)` |讀取JSON文件|
| `json.loads(<str>)` |將JSON字串轉成Python物件|

```python
# 物件轉JSON字串
import json

fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['greem', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] 
    }
}

with open('file.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(fruit_dict, indent = 4))
```

```python
# 寫入
with open('file.json', 'w', encoding='utf-8') as f:
    json.dump(fruit_dict,f , indent = 4)
```

```python
# 讀取
with open('file.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
print(data)
print(data['apple'])
```

```python
with open('file.json', 'r', encoding='utf-8') as f:
    texts = f.read()
    data = json.loads(texts) #把字串轉成該型別的資料 (只能是 list / dict)

print(data)
print(data['apple'])
```
