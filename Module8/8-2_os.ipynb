```python
# 匯入套件
import os

# 列出當前資料夾的檔案
files = os.listdir('../Module2')
print(files)

# ================================
# 檔案路徑
file_path = "file.txt"

print(os.path.exists(file_path))

# 如果指定路徑的檔案存在，則執行刪除
if os.path.exists(file_path):
    os.remove(file_path)
    print(f"{file_path} 已刪除")

```
