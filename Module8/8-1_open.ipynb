##### 讀取檔案 read( )

```python
# 舊寫法
file = open(file = "../Module2/2-1.py", mode = 'r', encoding="utf-8")

texts = file.read()

print(texts)

file.close() # 關閉該檔案
```

```python
# 印出 50 個 "="
print("=" * 50)

# 讀取檔案內容 - 常用的寫法 (執行完畢會自動關閉檔案)
# 只要一離開 with 語句 就關閉該檔案 __enter__, __exit__
with open(file = "../Module2/2-1.py", mode = 'r', encoding="utf-8") as f:
    texts = f.read()

print(texts)
# 印出 50 個 "="
print('=' * 50)

```

```python
# 一行一行讀取 把每一行放進 list
with open(file = "../Module2/2-1.py", mode = 'r', encoding="utf-8") as f:
    texts = f.readlines()

print(texts)
```

```python
# 資料處理
list = []
# file 與 mode 可以省略
with open("../Module2/2-1.py", 'r', encoding="utf-8") as f:
    for line in texts: # 這裡輸入 f 與 texts 都行
        list.append(line.strip())

print(list)
```

##### 寫入檔案 write( )

```python
# 文字內容
text = '''
傑出的人會努力規劃自己的人生，普通的人則是隨便命運怎麼安排   
一個是充實、有目的的過每一天，一個是隨興的過一天算一天
創業這條路 Michael E.Gerber
'''

# 寫入檔案
with open('file.txt', 'w', encoding='utf-8') as f:
    f.write(text)


```

```python
# 寫入資料到最後一行 (附加文字在檔案內容最後一行)
texts = "People say that success just happens to you. It doesn't!"
with open('file.txt', 'a', encoding='utf-8') as f:
    f.write(texts + '\n')
```
