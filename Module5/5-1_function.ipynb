```python
# 建立 function 1
# 沒有 return ， 也沒有參數
def greeting():
    print("Hello, World")

r = greeting()
print(r)
greeting()
greeting()
```

```python
# 建立 function 2
# 有 return 值
def hello_texts():
    return "Good Morning"

hello = hello_texts()
print(hello)
```

```python
# 建立 function 3
def hello_user(name):
    print(f"Hello, {name}")

hello_user("Ollie")
hello_user() # 會報錯，因為函式需要一個 name 參數，但呼叫時沒有傳入任何引數
```

```python
# 建立 function 4 # 預設參數值
# ❌def say_hello(greeting='Good Morning!', nqme):
def say_hello(name, greeting='Good Morning!'): # 要預設參數值的參數一定要在後面
    print(f"{greeting}, {name}")

# sya_hello() # 沒給參數會報錯
say_hello("Ollie")
say_hello("Ollie", "Hello!")
```

```python
# 改寫 Module2 practice
totals = [5600, 3400, 9000, 4000]

def get_discount(total: int, discount: int = 200) -> int:
    if total >= 2000:
        cnt = total // 2000 # 折幾個 200
        total -= discount * cnt

    return total

for total in totals:
    result = get_discount(total, discount = 100)
    print(result)
```
