```python
# 全域變數
name = "Ollie"

# 定義函式
def say_hello():
    name = "Alex" # 區域變數，不會影響到全域變數

    print(f"Hello {name}")

# 執行函式
say_hello()

# 輸出 name
print(name) # 沒有因為區域變數而改變
```

##### 使用 global 讓函式存取全域變數的值

```python
# 全域變數
name = "Ollie"

# 定義函式
def say_hello():
    global name
    name = "Alex" # 對全域變數做重新賦值

    print(f"Hello {name}")

# 執行函式
say_hello()

# 輸出 name
print(name) # 因為有呼叫 global 而改變
```
